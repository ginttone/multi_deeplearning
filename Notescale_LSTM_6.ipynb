## 데이터 전처리

1차원 -> 2차원으로 변경

```python
note_seq = ['g8', 'e8', 'e4', 'f8', 'd8', 'd4', 'c8', 'd8', 'e8', 'f8', 'g8', 'g8', 'g4',
       'g8', 'e8', 'e8', 'e8', 'f8', 'd8', 'd4', 'c8', 'e8', 'g8', 'g8', 'e8', 'e8', 'e4',
       'd8', 'd8', 'd8', 'd8', 'd8', 'e8', 'f4', 'e8', 'e8', 'e8', 'e8', 'e8', 'f8', 'g4',
       'g8', 'e8', 'e4', 'f8', 'd8', 'd4', 'c8', 'e8', 'g8', 'g8', 'e8', 'e8', 'e4']
```

```python
note_seq[0:5], note_seq[1:6], note_seq[2:7]
```

```python
code2idx = {'c4':0, 'd4':1, 'e4':2, 'f4':3, 'g4':4, 'a4':5, 'b4':6,
            'c8':7, 'd8':8, 'e8':9, 'f8':10, 'g8':11, 'a8':12, 'b8':13}
```

```python
len(note_seq), range(len(note_seq)-5)
```

```python
dataset = list()
for i in range(len(note_seq)-5):
  subset = note_seq[i:i+5]
  items = list()
  for item in subset:
     items.append(code2idx[item])
  dataset.append(items)

print(dataset)
```

```python
import numpy as np
```

```python
datasets = np.array(dataset)
```

```python
x_train = datasets[:,0:4]
x_train.shape
```

```python
y_train = datasets[:,3]
y_train.shape
```

```python
len(code2idx)
```

```python
x_train = x_train/14
x_train[3]
```

```python
X_train = np.reshape(x_train, (49, 4, 1))
X_train.shape, X_train[2]
```

```python
np.unique(y_train)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.Input(shape=(4,1))) 
model.add(tf.keras.layers.LSTM(128)) 
model.add(tf.keras.layers.Dense(13, activation='softmax')) 

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
hist = model.fit(X_train, y_train, epochs=1000, batch_size=10)
```

```python
model.evaluate(X_train,y_train)
```
